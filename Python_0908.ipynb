## 2. 조건문

```python
# if
# 조건이 끝날 때 콜론(:) 주의, 들여쓰기 주의
# 조건문은 결과값이 True or False가 되어야 한다. 
# 조건문의 결과가 True야만 조건문은 실행된다.

people = 3
apple = 32

if people < apple / 5 :
  print("사과파티 !!!")

if apple % people > 0 :
  print("나머지는 나눠 먹자 !!")

if people > apple :
  print("몇 명은 못먹겠네...ㅠㅠ")     # if를 연달아 사용해도 되긴 하지만, 첫 조건을 만족해도 두번째, 세번째 조건을 계속 확인한다.
```

```python
# 가위바위보 
scissor = "가위"
rock = "바위"
paper = "보"

win = "이겼다 !"
draw = "비겼다 !"
lose = "졌다.. ㅠㅠ"

mine = "보"
your = "보"

if mine == your :
  result = draw
else :
  if mine == scissor :
    if your == rock :
      result = lose
    else : 
      result = win
  else :
    if mine == rock :
      if your == paper :
        result = lose
      else :
        result = win
    else :
      if mine == paper :
        if your == scissor :
          result = lose
        else :
          result = win


print(result)
```

## 3. 함수

```python
# 기능, 클래스 안에서는 메소드 라고 부른다.

a = 1
b = 2
c = -8

# x에 관한 2차방정식 
# a * x^2 + b * x + c (단, a != 0)

r1 = (-b + (b**2 -4*a*c) ** 0.5) / (2 * a) # 근의공식
r2 = (-b - (b**2 -4*a*c) ** 0.5) / (2 * a)

print(r1, r2)
```

```python
# 근의공식을 사용할 때마다 위처럼 사용하기 번거로우니 함수로 재정의 한다 -> 재사용이 매우 편하다.
def print_root(a,b,c):
  r1 = (-b + (b**2 -4*a*c) ** 0.5) / (2 * a)
  r2 = (-b - (b**2 -4*a*c) ** 0.5) / (2 * a)

  print("해는 {} 또는 {}".format(r1,r2))

print_root(1,2,1)
```

```python
# 리스트는 여러 자료를 한번에 담아서 꺼내 쓴다. 관리하기가 편리하다.

list1 = ['가위','바위','보']
list2 = [27,23,32,74,45,32,21]

print(list1[2])
# print(list1[5]) # 인덱스 범위에 없으므로 IndexError: list index out of range
```

```python
# 멤버 이름에 번호 부여하기

bts = ['RM', '진', '슈가','제이홉','지민','뷔','정국']
for i in range(7):
  name = bts[i]
  print('{}번:{}'.format(i+1,name),end = ' ')

print()
# 위에 코드는 리스트를 다 세어봐야하는 번거로움이 있다.. len() 함수를 이용
bts = ['RM', '진', '슈가','제이홉','지민','뷔','정국','유나']
for i in range(len(bts)):
  name = bts[i]
  print('{}번:{}'.format(i+1,name), end = ' ')

print()
# 조금 더 업그레이드... enumerate() 함수를 이용 : 하나씩 번호를 부여해준다.
bts = ['RM', '진', '슈가','제이홉','지민','뷔','정국','유나']
for i, name in enumerate(bts):
  print('{}번:{}'.format(i+1,name), end = ' ')

print()
# enumerate() 함수 실제 모습, 튜플로 넘버링을 해준다 !!!
for i in enumerate(bts) :
  print(i, end=' ')
```

## 8. 딕셔너리 / 튜플


```python
# 딕셔너리 : key값과 value값으로 구성, 키로 밸류 값을 조회 {   } -> 인덱싱과 슬라이싱 불가능, 수정/삭제 가능 mutable
wintable = {
    '가위' : '보',
    '바위' : '가위',
    '보' : '바위'
}

print(wintable['가위'])
```

```python
# if에서 만들었던 가위바위보 게임을 딕셔너리로 간단하게 코딩
# 이기는 경우의 수를 딕셔너리 형태로 구성
wintable = {
    '가위' : '보',
    '바위' : '가위',
    '보' : '바위'
}

def rsp(mine,yours) :
  if mine == yours :
    return '비겼다!'
  elif wintable[mine] == yours :
    return "내가 이겼다 !"
  else :
    return "내가 졌다..ㅠㅠ"

print(rsp('가위','바위'))
```

```python
# 딕셔너리 수정하기

# 먼저 리스트 수정해보기 
list = [1,2,3,4,5]
list[2] = 33    # 인덱스로 수정하기
print(list)

#list[5] = 66  # IndexError: list assignment index out of range
#print(list)  

list.append(66)
print(list) 

# 리스트의 삭제
del(list[0])  # 역시 인덱스로 삭제
print(list)

# 또는 pop() 활용
list.pop(0)   # 인덱스 0번째꺼 삭제
print(list)
```

```python
# 그럼 딕셔너리를 수정해보자
dict = {'one' : 1, 'two' : 2, 'three' : 3}
dict['two'] = 22   # 키값으로 수정한다
print(dict)

# 딕셔너리를 추가 하고싶은데 list처럼 append가 없다! 그래서
dict['four'] = 4    # 키값을 새로 지정해주고 밸류를 넣으면 된다. 대신 순서가 없다...
print(dict)

# 딕셔너리 삭제
del(dict['one'])   # 키값을 입력해서 키-밸류 같이 삭제
print(dict)

# 딕셔너리도 리스트의 pop()처럼 삭제 가능하나 인덱스(순서)가 없으므로 key값을 입력
dict.pop('four')
print(dict)
```

## 9. while

```python
# while은 반복 횟수를 모를 때 주로 사용 ! 

select = None
while select not in ['가위','바위','보'] :
  select = input('가위,바위,보 중에서 선택하세요 >>')

print('선택한 값은 : ' ,select)
```

```python
# 만약 if문으로 한다면? 만족할때까지 돌지 않기 때문에 그냥 한번에 끝나버린다...
select = None

if select not in ['가위','바위','보'] :
  select = input('가위,바위,보 중에서 선택하세요 >>')

print('선택한 값은 : ' ,select)
```

```python
# for문과 while문의 비교 -> 값이 똑같게 코드 짜보기
pattern = ['가위','바위','보']

# for문으로 한다면? 

for i in range(len(pattern)) :
  print(pattern[i])

print()

# while문으로 바꾸면?
i = 0
while i < len(pattern) :
  print(pattern[i])
  i += 1 
```
